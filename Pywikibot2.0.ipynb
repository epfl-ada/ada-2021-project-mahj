```python
df = pd.read_csv(FILE_NEWSPAPER_SPEAKER_COUNT,compression='bz2')
```

```python
# The pbm is that if cannot find page, search for pages containing keyword and pick first one
# => must get dump and look at alias and website

owned_by_key = 'P127'
site = pywikibot.Site('en', 'wikipedia')
repo = site.data_repository()

def extract_QID_owner(newspaper):
    
    page = pywikibot.Page(site, newspaper)
    QID = None
    owner = None
    if(page.exists()):
        if(page.isRedirectPage()):
            page = page.getRedirectTarget()
        item = page.data_item()
        QID = item.getID()
        claims = item.get()['claims'].toJSON()# Provides dict["property_key"->property]
        owned_by_property = claims.get(owned_by_key)#Return None if property not found
        if(owned_by_property!=None):
            owner_id = owned_by_property[0]['mainsnak']['datavalue']['value']['numeric-id']

            #Use ItemPage to obtain a dictionnary
            data_found = pywikibot.ItemPage(repo, "Q" + str(owner_id))
            #Get title in english
            owner = data_found.get()['labels']['en']

    return QID,owner
```

```python
newspaper_owners = {}
for i,newspaper in enumerate(df['newspaper'].unique()):
    if(i%10 == 0):
        print(f'Iteration {i}')
    print(newspaper)
    QID,owner = extract_QID_owner(newspaper)
    newspaper_owners[newspaper] = (QID,owner)
    print(QID,owner)
```
